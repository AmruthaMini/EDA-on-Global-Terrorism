## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

**Number of terror attacks over past five decades**

````python
# Chart - 1 visualization code
gtd_df1 = df2.groupby('year')['eventid'].count().reset_index().rename(columns={'eventid':'no_of_attacks'})
````

````python
gtd_df1
````

````python
plt.rcParams["figure.figsize"]=(20,10)
sns.lineplot(data = gtd_df1
            ,x = 'year'
            ,y = 'no_of_attacks'
            )

plt.title("Trend in number of terrorist attacks over the past five decades",fontsize=25)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Number of attacks",fontsize=20)
plt.show()
````

##### 1. Why did you pick the specific chart?

A line chart is best to visualise change over time. It gives track over a period of time.

##### 2. What is/are the insight(s) found from the chart?

After 2010, the number of terror attacks increased and 2014 witnessed the highest over the past five decades. This surge ref;ects heightened global instability. Later after 2015, there is a downward trend in the number of terror attacks. This decline suggests the potential effectivenesss of international counter-terrorism strategies.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Helps governments replicate successful security startegies and avoid ineffective ones.

#### Chart - 10

**Trend in number of terrorist attacks,deaths and wounded civilians over the past five decades**

````python
# Chart - 10 visualization code
tc = df2.groupby('year').agg({'eventid':"count",'deaths':'sum',"injured":'sum'})
tc.rename(columns={'eventid':"num_of_attacks",'deaths':"num_of_civilians_killed","injured":"num_of_civilians_wounded"},inplace=True)
tc.reset_index(inplace=True)
tc.head()
````

````python
plt.rcParams["figure.figsize"]=(25,10)
tc.plot(x="year",y=['num_of_attacks','num_of_civilians_killed','num_of_civilians_wounded'],kind="area",stacked=False).legend(fontsize=20)
plt.title("Trend in number of terrorist attacks,deaths and wounded civilians over the past five decades",fontsize=25)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Count",fontsize=20)

plt.show()
````

##### 1. Why did you pick the specific chart?

Since area chart is a graph that combines a line chart and a bar chart to show changes in quantities over time.
